```python
source_dir = "/Volumes/incremental_load/default/orders_data/source/"
target_dir = "/Volumes/incremental_load/default/orders_data/archive/"
stage_table = "incremental_load.default.orders_stage"
```

```python
# df = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load(source_dir)
df = spark.read.csv(source_dir, header=True, inferSchema=True)
# df = spark.read.option("header", "true").option("inferSchema", "true").csv("gs://incremental_load_data_here/source/order_tracking_2023_01_01.csv")
```

```python
df.write.format("delta").mode("overwrite").saveAsTable(stage_table)
```

```python
files = dbutils.fs.ls(source_dir)

for file in files:
    # print("file", file)
    source_path = file.path
    # print("file.path: ",file.path)
    # print("source_path: ",source_path)
    target_path = target_dir + source_path.split("/")[-1]
    # print("operation: ", source_path.split("/")[-1])
    # print("target_path: ",target_path)
    dbutils.fs.mv(source_path, target_path)
```
